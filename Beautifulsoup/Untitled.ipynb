```python
import requests

from bs4 import BeautifulSoup
```

```python
page_count = 1

while True:
    URL  = "https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(URL)

    soup = BeautifulSoup(res.text,"lxml")
    quotes = soup.select("div.quotes")

    if not quotes:
        print("invalid page")
        break

    with open(f"scraped_data/quotes{page_count}.html","w") as f:
        f.write(res.text)
        print(f"scraped_data/quotes{page_count}.html")
        
    page_count = page_count + 1
```
